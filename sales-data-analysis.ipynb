## Handling Missing Values

Some columns in the dataset contain missing (null) values.
These columns are related to customer address and region information.
Instead of deleting rows, we fill these missing values using appropriate methods to preserve important sales data.

```python
df['ADDRESSLINE2']= df['ADDRESSLINE2'].fillna("Not Provided")
df['STATE'] = df['STATE'].fillna(df['STATE'].mode()[0])
df['POSTALCODE']=df['POSTALCODE'].fillna('Unknown')
df['TERRITORY']=df['TERRITORY'].fillna(df['TERRITORY'].mode()[0])
```

```python
df.isnull().sum()
```

## Creating Total Sales Column
We calculate total sales for each order using quantity ordered and price per unit.
This helps in analyzing revenue.

```python
df['SALES']=df['QUANTITYORDERED']*df['PRICEEACH']
```

## Overall Sales Performance

```python
total_sales = df['SALES'].sum()
total_sales
```

```python
df.columns
```

## Year-wise Sales Trend 
This chart show haw total sales have changed across different years.
It helps in understanding business growth or decline over time.

```python
year_sales=df.groupby("YEAR_ID",as_index = False)['SALES'].sum()


plt.figure(figsize=(8,5))
plt.bar(year_sales['YEAR_ID'],year_sales['SALES'])
plt.title("Year-wise Sales")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

## Monthly Sales Trend
This analysis show how sales vary across different months of year.
It helps identify seasonal patterns,peak months, and low-performing months, which is useful for plannong marketing campaigns and inventory management.

```python
month_sales= df.groupby('MONTH_ID',as_index=False)['SALES'].sum()
month_sales

plt.figure(figsize=(8,5))
plt.plot(month_sales['MONTH_ID'],month_sales['SALES'])
plt.title("Monthly Sales Trend")
plt.xlabel("Months")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

## Sales by Product Line
This chart show which product catagories contribute the most to total sales.

```python
product_sales = df.groupby('PRODUCTLINE',as_index=False)['SALES'].sum().sort_values(by='SALES',ascending=False)

plt.figure(figsize=(8,5))
plt.barh(product_sales['PRODUCTLINE'],product_sales['SALES'])
plt.title("Sales by Product Line")
plt.xlabel("Total Sales")
plt.ylabel("Product Line")
plt.show()
```

## Top 10 Customers by Sales
This graph shows the customers who generate the highst revenue.

```python
top_customers = df.groupby('CUSTOMERNAME',as_index=False)['SALES'].sum().sort_values(by='SALES',ascending=False).head(10)

plt.figure(figsize=(8,5))
plt.bar(top_customers['CUSTOMERNAME'],top_customers['SALES'])
plt.title('Top 10 Customers by Sales')
plt.xticks(rotation=45)
plt.ylabel("Total Sales")
plt.show()
```

## Top 10 Countries
This chart shows the countries that contribute most to total sales.

```python
country_sales= df.groupby('COUNTRY',as_index=False)['SALES'].sum().sort_values(by='SALES',ascending=False).head(10)

plt.figure(figsize=(8,5))
plt.bar(country_sales['COUNTRY'],country_sales['SALES'])
plt.title("Top 10 Country by Sales")
plt.ylabel("Total Sales")
plt.show()
```
